```python
import numpy as np
import pandas as pd
import statsmodels.api as sm
import matplotlib.pyplot as plt
import seaborn as sns
sns.set
```

```python
raw_data = pd.read_csv('Book3.csv')
```

```python
# .copy:this is to copy the value of dummy to data
#.map:is used to map a value to a variable
data = raw_data.copy()
data['average'] = data['average'].map({'yes':1, 'no':0})
data
```

```python
y = data['average']
x1 = data['scores']
```

```python
x = sm.add_constant(x1)
reg_log = sm.Logit(y,x)
results_log = reg_log.fit()
```

```python
results_log.summary()
```
