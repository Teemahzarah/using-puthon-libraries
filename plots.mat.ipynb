```python
import numpy as np
import pandas as pd 
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
x =np.linspace(0,5,11)
y = x**2
```

```python

x,y
```

```python
plt.plot(x,y)
plt.xlabel('numbers')
plt.ylabel('sqrs of numbers')
plt.title('title')
plt.show()#u dont necsarily need this function if ure using j_notebook
```

```python
plt.subplot(1,2,1)
plt.plot(x,y,'b')

plt.subplot(1,2,2)
plt.plot(x,y,'r')
```

```python
fig =plt.figure()
axes1 =fig.add_axes([0.1,0.1,0.8,0.8])
axes2 = fig.add_axes([0.4,0.4,0.5,0.4])
axes1.plot(x,y)
axes2.plot(y,x)
```

```python
fig,axes =plt.subplots(nrows=1, ncols=2)

axes[0].plot(x,y)
axes[0].set_title('A')

axes[1].plot(y,x)
axes[1].set_title('B')
```

```python
fig =plt.figure()
axe =fig.add_axes([0,0,1,1])

axe.plot(x,x**2,label='squre',color='#ff8c00',linewidth=2,
        alpha=0.5)
axe.plot(x,x**3,label='cube',color='purple',lw=1,
         linestyle='steps',marker='o')
#we can also specify markersize,markerfacecolor,
#makeredgewidth...
axe.legend()
```
